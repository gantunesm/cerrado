# View all cerrado colormaps

```python
import matplotlib.pyplot as plt
import numpy as np
import cerrado
 
```

```python
cerrado.view_all_colors()
```

# Default color cycles

The default of cerrado is 'chuveirinho'

```python
a = np.linspace(0, 2*np.pi, 100)

plt.plot(a, np.sin(a))

plt.plot(a, np.cos(a))
plt.plot(a, np.cos(a+0.5))
plt.plot(a, np.cos(a+1))
plt.plot(a, np.cos(a+1.5))
plt.plot(a, np.cos(a+2))

plt.title('Title')
plt.xlabel('axis-x')
plt.ylabel('axis-y')
```

### Picking another color cycle

### Picking another colormap

```python
cerrado.set_colormap('cer2')
plt.imshow(Z)
```

```python

def func3(x, y):
    return (1 - x / 2 + x**3 + y**2) * np.exp(-(x**2 + y**2))


# make these smaller to increase the resolution
dx, dy = 0.03, 0.03

x = np.arange(-3.0, 3.0, dx)
y = np.arange(-3.0, 3.0, dy)
X, Y = np.meshgrid(x, y) 
Z2 = func3(X, Y)
```

```python
cerrado.set_colormap('santa_barbara')
plt.imshow(Z2)
plt.colorbar()
```

```python
cerrado.set_colormap('cer2')
plt.imshow(Z2)
plt.colorbar()
```

```python
cerrado.set_colormap('laelia')
plt.imshow(Z2)
plt.colorbar()
```
