# View all cerrado colormaps

```python
import matplotlib.pyplot as plt
import numpy as np
import cerrado
```

```python
cerrado.view_all_colors()
```

# Default color cycles

The default of cerrado is 'cer1'

```python
a = np.linspace(0, 2*np.pi, 100)

plt.plot(a, np.sin(a))

plt.plot(a, np.cos(a))
plt.plot(a, np.cos(a+0.5))
plt.plot(a, np.cos(a+1))
plt.plot(a, np.cos(a+1.5))
plt.plot(a, np.cos(a+2))

plt.title('Title')
plt.xlabel('axis-x')
plt.ylabel('axis-y')
```

### Picking another color cycle

```python
cerrado.set_color_cycle('cer2')
```

```python
a = np.linspace(0, 2*np.pi, 100)

plt.plot(a, np.sin(a))

plt.plot(a, np.cos(a))
plt.plot(a, np.cos(a+0.5))
plt.plot(a, np.cos(a+1))
plt.plot(a, np.cos(a+1.5))
plt.plot(a, np.cos(a+2))

plt.title('lalala')
plt.xlabel('kkkkkkk')
plt.ylabel('kkkkkkk')
```
